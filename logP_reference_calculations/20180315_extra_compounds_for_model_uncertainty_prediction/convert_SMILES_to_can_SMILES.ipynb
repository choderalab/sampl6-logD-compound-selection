```python
from openeye.oechem import *
import pandas as pd
```

```python
# Import table with OeLexigem generated SMILES (nam2mol_example.py)

df = pd.read_csv("extra_molecule_names_and_SMILES.csv")
df["Canonical SMILES"] = None
df
```

```python
# Write canonical isomeric SMILES to a new column
for i in range(df.shape[0]):
    smiles = df.loc[i, "SMILES"]
    mol = OEGraphMol()
    OESmilesToMol(mol, smiles)
    canonical_smiles = OEMolToSmiles(mol)
    df.loc[i,"Canonical SMILES"] = canonical_smiles
 
df
```

```python
# Combine together with Table 1

df_table1 = pd.read_csv("extra_molecules_with_pHmetric_logP_only_neutral.csv")
df_table1
```

```python
df_table1["Canonical SMILES"] = None

# Iterate through each entry to add canonical isomeric SMILES to the table

for i in range(df_table1.shape[0]):
    canonical_smiles = df.loc[i, "Canonical SMILES"]
    df_table1.loc[i, "Canonical SMILES"] = canonical_smiles
    
df_table1
```

```python
# Write to a CSV
df_table1.to_csv("extra_molecules_with_pHmetric_logP_only_neutral_with_canonical_SMILES.csv", index=False)
```
